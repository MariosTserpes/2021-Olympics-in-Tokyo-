* Let's see how the athletes are distributed by nationality : TOP 10

```python
no_of_comp_by_country = pd.DataFrame(athletes.groupby(['NOC'])['Discipline'].count().sort_values(ascending = False)[:10]).T
no_of_comp_by_country.columns = ['USA', 'Japan', 'Australia', 'China', 'Germany', 'France', 'Canada', 'G. Britain', 'Italy', 'Spain']
```

```python
sum_df = no_of_comp_by_country.sum().to_frame().T
cols = no_of_comp_by_country.columns.to_list()
title = '2021 Olympics in Tokyo. Athletes by Nationality. TOP10 countries'

fig, ax = plt.subplots(1, len(cols), figsize=(17, 4))
ax = ax.flatten()

ax[0].text(0.02, 0.9, title, size = 22, color = 'black', weight='bold')
ax[0].text(0.02, 0.7, f"Total Number of Athletes for Top 10 countries is : {np.sum(no_of_comp_by_country.values)}",
           size = 15, color = 'blue', fontweight = 'semibold')

for ind, col in enumerate(cols):
    ax[ind].text(0.5,
                 0.5,
                 col.title(),
                 ha='center',
                 va='center',
                 fontsize=17,
                 color='lightblue',
                 weight='heavy',
                 bbox=dict(edgecolor = 'yellow',
                           facecolor= 'black',
                           lw=2,
                           pad=10))
    
    ax[ind].text(0.5,
                 0.3,
                 f"{int(sum_df[col])}\n ({round(100 * int(sum_df[col]) / (np.sum(no_of_comp_by_country.values)), 1)}%)", 
                 ha='center',
                 va='center',
                 fontfamily='monospace',
                 fontsize=20,
                 fontweight='bold',
                 color='blue')
    
    ax[ind].set_axis_off()
```
